```python
import numpy as np
import pandas as pd
#import matplotlib to plot chart
import matplotlib.pyplot as plt
#plot inline in notebook (not separate page)
%matplotlib inline
```

```python
content = pd.read_csv('TidyDataSheet.csv')
```

```python
content.head()
```

```python
#filtering data for voiced
voiced = content[content.voicing == 'Vd']
voiced
#filtering data for Vd fast
voiced_fast = voiced[voiced.speed == 'fast']
Vd_fast_avg = voiced_fast['duration'].mean()
#filtering data for Vd normal
voiced_normal = voiced[voiced.speed == 'normal']
Vd_normal_avg = voiced_normal['duration'].mean()
#filtering data for Vd slow
voiced_slow = voiced[voiced.speed == 'slow']
Vd_slow_avg = voiced_slow['duration'].mean()
#filtering data for voiceless
voiceless = content[content.voicing == 'Vl']
voiceless
#filtering data for Vl fast
voiceless_fast = voiceless[voiceless.speed == 'fast']
Vl_fast_avg = voiceless_fast['duration'].mean()
#filtering data for Vl normal
voiceless_normal = voiceless[voiceless.speed == 'normal']
Vl_normal_avg = voiceless_normal['duration'].mean()
#filtering data for Vl slow
voiceless_slow = voiceless[voiceless.speed == 'slow']
Vl_slow_avg = voiceless_slow['duration'].mean()

#plotting averages
Voiced_avg = {"Voiced": ["fast", "normal", "slow"], "avg":
             [Vd_fast_avg, Vd_normal_avg, Vd_slow_avg]}
Voiced_avgdf = pd.DataFrame(Voiced_avg)

import seaborn as sns
sns.set()

plt.figure(figsize = (10,6))
plt.suptitle("Avg. vowel length preceding voiced/voiceless consonants by speech rate",
fontsize = 20)
#plt.supylabel("Speech Rate")
plt.subplot(1,3,1)
plt.bar("Voiced", Vd_fast_avg)
plt.bar("Voicless", Vl_fast_avg)
plt.title("Fast Speech Rate")
plt.ylim([0,250])
plt.ylabel("Average Vowel Duration [ms]", fontsize = 20)

plt.subplot(1,3,2)
plt.bar("Voiced", Vd_normal_avg)
plt.bar("Voiceless", Vl_normal_avg)
plt.title("Normal Speech Rate")
plt.ylim([0,250])

plt.subplot(1,3,3)
plt.bar("Voiced", Vd_slow_avg)
plt.bar("Voiceless", Vl_slow_avg)
plt.title("Slow Speech Rate")
plt.tight_layout()
plt.ylim([0,250])
```

```python
plt.figure(figsize = (10,4))
plt.title("Difference between avg. vowel length preceding voiced vs voiceless consonants", fontsize = 20)
#plt.supylabel("Speech Rate")
plt.bar("Fast", Vd_fast_avg - Vl_fast_avg)
plt.bar("Normal", Vd_normal_avg - Vl_normal_avg)
plt.bar("Slow", Vd_slow_avg - Vl_slow_avg)
plt.xlabel("Speech Rate", fontsize = 16)
plt.ylabel("Difference in avg. vowel duration (ms)", fontsize = 12)
plt.tight_layout()
```
